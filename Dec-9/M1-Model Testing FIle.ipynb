```python
def drop_ManyPrices(df1, df2, onlyDropHalf, saveAverage):
  if onlyDropHalf:
    if saveAverage:
      df1.drop(['MMRAcquisitionAuctionCleanPrice', 'MMRAcquisitonRetailCleanPrice', 'MMRCurrentAuctionCleanPrice', 'MMRCurrentRetailCleanPrice'],axis=1,inplace=True)
      df2.drop(['MMRAcquisitionAuctionCleanPrice', 'MMRAcquisitonRetailCleanPrice', 'MMRCurrentAuctionCleanPrice', 'MMRCurrentRetailCleanPrice'],axis=1,inplace=True)
    else:
      df1.drop(['MMRAcquisitionAuctionAveragePrice','MMRAcquisitionRetailAveragePrice','MMRCurrentAuctionAveragePrice','MMRCurrentRetailAveragePrice'],axis=1,inplace=True)
      df2.drop(['MMRAcquisitionAuctionAveragePrice','MMRAcquisitionRetailAveragePrice','MMRCurrentAuctionAveragePrice','MMRCurrentRetailAveragePrice'],axis=1,inplace=True)
  else:
    df1.drop(['MMRAcquisitionAuctionAveragePrice','MMRAcquisitionRetailAveragePrice','MMRCurrentAuctionAveragePrice','MMRCurrentRetailAveragePrice', 'MMRAcquisitionAuctionCleanPrice', 'MMRAcquisitonRetailCleanPrice', 'MMRCurrentAuctionCleanPrice', 'MMRCurrentRetailCleanPrice'],axis=1,inplace=True)
    df2.drop(['MMRAcquisitionAuctionAveragePrice','MMRAcquisitionRetailAveragePrice','MMRCurrentAuctionAveragePrice','MMRCurrentRetailAveragePrice', 'MMRAcquisitionAuctionCleanPrice', 'MMRAcquisitonRetailCleanPrice', 'MMRCurrentAuctionCleanPrice', 'MMRCurrentRetailCleanPrice'],axis=1,inplace=True)
```

```python
numerical_features = [c for c, dtype in zip(X.columns, X.dtypes)
                     if dtype.kind in ['i','f']]
categorical_features = [c for c, dtype in zip(X.columns, X.dtypes)
                     if dtype.kind not in ['i','f']]
```

```python
#import train_test_split library
from sklearn.model_selection import train_test_split

# create train test split
X_train, X_test, y_train, y_test = train_test_split( X,  y, test_size=0.4, random_state=0, stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.impute import KNNImputer
from sklearn.impute import IterativeImputer
# imputer=SimpleImputer(strategy='mean');
# imputer = IterativeImputer(random_state=4)

preprocessor = make_column_transformer(

        # (make_pipeline(
        # SimpleImputer(strategy = 'median'),
        # StandardScaler()), numerical_features),

        # (make_pipeline(
        # KNNImputer(n_neighbors=20),
        # StandardScaler()), numerical_features),

        (make_pipeline(
        imputer = IterativeImputer(random_state=3),
        StandardScaler()), numerical_features),


        (make_pipeline(
        SimpleImputer(strategy = 'constant', fill_value = 'missing'),
        OneHotEncoder(categories = 'auto', handle_unknown = 'ignore')), categorical_features),

)
from sklearn.neural_network import MLPRegressor
# Model = make_pipeline(preprocessor, LogisticRegression(class_weight="balanced", solver= "sag"))
Model = make_pipeline(preprocessor, MLPRegressor(max_iter=200))
# learning_rate='adaptive
```
